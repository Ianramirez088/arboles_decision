```python
# Definimos cuales son las variables que trbajaremos en las predcciones, y cual es la variable objetivo
predictors_col = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "total sulfur dioxide", "density", "pH", "sulphates", "alcohol"]
target_col = ["quality"]
```

```python
# Definimos las variables con las que haremos las predicciones, y variable objetivo, y les asignamos los valores del dataset
predictors = data[predictors_col]
target = data[target_col]
```

```python
# Imprimimos en pantalla la variable con la que trabajaremos
predictors
```

```python
# Imprimimos en pantalla la variable objetivo
target
```

```python
corr = predictors.corr()
corr
```

**Mapa de calor**

```python
sns.heatmap(corr)
plt.show()
```

**Correlación de variables**
Las acides del vino depende del acido citrico y acidez fija que compone el vino, es decir que a mayor acido citrico en el vino, mayor acidez fija tiene el vino, y sucede lo mismo en sentido inverso.
El total de dióxido de azufre, aumenta proporcionamente con respecto al dióxido de azufre libre.

**Entrenamiento de datos**

```python
# Entrenamos el algoritmo con un porcentaje de entrenamiento del 78%
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.22, random_state=13)
```

```python
tree = DecisionTreeClassifier()
```

**Entrenar el modelo configurando los diferentes 
hiperparámetros.**

```python
# Preparamos el arbol de desición con los datos de entrenamiento
arbol = tree.fit(X_train, y_train)
```

```python
# Mostramos el arbol de decision
plot_tree(arbol)
```

```python
# Realizamos las predicciones con el 22% de los datos faltantes, que son los de prueba
predicciones = arbol.predict(X_test)
```

```python
# Mostramos las predicciones
predicciones
```

**Evaluar el desempeño del modelo en el conjunto de Test con 
métricas como precisión, recall, F1-score, etc.**

```python
# Examinamos matriz de confusión
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=['Actual'], colnames=['Predicciones'])
```

```python
# Aproximación de predicciones con valores reales
accuracy = accuracy_score(y_test, predicciones)
accuracy
```

**Conclusión:** Como podemos notar, la certeza de la predicción del algoritmo, no es muy alta, por lo cual no se podría confiar mucho en la fiabilidad del algoritmo, para saber la calidad del vino, con el 22% de los datos restantes a los del entrenamiento, se debe mejorar el algoritmo para una predicción más precisa.
